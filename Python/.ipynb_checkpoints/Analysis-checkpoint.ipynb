# Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from itertools import combinations 
```

# Retention & Cohort Analysis

## Retention

### With Ratio

```python
cohort_ratio=pd.read_csv('./Data/cohort_table_ratio.csv', index_col='first_order')
cohort_ratio
```

```python
plt.rcParams['axes.facecolor'] = 'white'
fig, ax = plt.subplots(figsize = (15, 8))

mask = np.zeros_like(cohort_ratio.corr(method = 'pearson'), dtype=np.bool)
mask[np.triu_indices_from(mask,k=1)] = True
mask = np.flip(mask,axis=0)

sns.heatmap(cohort_ratio, cmap = 'Greys', annot = True,
            fmt = '.2%', annot_kws = {'fontsize': 12}, cbar = False,
            ax = ax, mask=mask)


plt.yticks(rotation = 0)
plt.show()
```

### With Value

```python
cohort=pd.read_csv('./Data/cohort_table.csv', index_col='first_order')
cohort
```

```python
cus_diff={'month_gap':[c for c in cohort.columns],
          'customer_gap':[sum(cohort[m]) for m in cohort.columns]}
cus_df=pd.DataFrame(cus_diff)
cus_df.set_index('month_gap',inplace=True)
cus_df
```

```python
plt.rcParams['axes.facecolor'] = 'white'
fig, ax = plt.subplots(figsize = (15, 8))
ax.grid(False)
x = cus_df.index[1:]
y = 100*(cus_df['customer_gap'][1:] / cus_df['customer_gap'][0])

sns.lineplot(x = x, y = y, color = '#595959', marker = 'o')

for x, y in zip(x, y):
    plt.text(x, y +0.5, s = round(y, 2))

plt.xlabel('Month Diff')
plt.ylabel('Customer Diff Rate')
plt.ylim(0,20)
sns.despine()
plt.show()
```

# Clustering

## Draw Scatter Plot

```python
sub_ca = pd.read_csv('./data/sub_ca.csv')
sub_ca
```

```python
x = sub_ca['quantity']
y = sub_ca['pbyq']
for i in range(len(sub_ca)):
    plt.scatter(x[i], y[i],label=sub_ca['Sub-Category'][i])


plt.xlabel('Quantity')
plt.ylabel('Total Profit by Quantity')
plt.title('Visualization by 2 Columns')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
plt.show()
```

## Clustering With KMeans

### Normalization

```python
sub_ca_a=sub_ca[['quantity','pbyq']]
scaler = MinMaxScaler(feature_range=(0, 1))
Dataset = scaler.fit_transform(sub_ca_a)
sub_ca_a = pd.DataFrame(Dataset, index=sub_ca_a.index,
                              columns=sub_ca_a.columns)

sub_ca_a
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300,random_state=0)
kmeans.fit(sub_ca_a)

sub_ca_a['cluster']=kmeans.labels_
sub_ca.insert(4,'Cluster',sub_ca_a['cluster'])
sub_ca_a
```

```python
marker_idx=[sub_ca_a[sub_ca_a['cluster']==i].index for i in range(3)]
marker={0:'o',1:'s',2:'+',3:'v'}
label=['Group0','Group1','Group2','Group3']

for i in range(3):
    plt.scatter(x=sub_ca_a.loc[marker_idx[i],'quantity'],
                y=sub_ca_a.loc[marker_idx[i],'pbyq'],
                label=label[i],
                marker=marker[i]) 

plt.xlabel('Quantity')
plt.ylabel('Total Profit by Quantity')
plt.title('2 Group Visualization by 2 Columns')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Association Analysis With Apriori

```python
sold_product=pd.read_csv('./Data/sold_product.csv')
sold_product
```

## Generate Dataset For Apriori

```python
for i in range(len(sold_product)):
    sold_product.at[i,'sold_product']=sold_product.loc[i,'sold_product'].split(',')
sold_product
```

## Apriori

```python
te = TransactionEncoder()
te_ary = te.fit(sold_product['sold_product']).transform(sold_product['sold_product'])
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.0000000001, use_colnames=True)
frequent_itemsets
```

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
```
