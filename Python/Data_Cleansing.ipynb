## Cleansing DataFrame Format

```python
for i in range(len(product)):
    if pd.notnull(product.loc[i]['b']):
        product.loc[i,'Product Name']=product.loc[i,'Product Name']+product.loc[i,'Sales']+product.loc[i,'Quantity']
        product.loc[i,'Sales']=product.loc[i,'Quantity']
        product.loc[i,'Quantity']=product.loc[i,'Discount']
        product.loc[i,'Discount']=product.loc[i,'Profit']
        product.loc[i,'Profit']=product.loc[i,'a']
        product.loc[i,'a']=product.loc[i,'b']
product=product.drop(labels='b',axis=1)
        
```

```python
for i in range(len(product)):
    if pd.notnull(product.loc[i]['a']):
        product.loc[i,'Product Name']=product.loc[i,'Product Name']+product.loc[i,'Sales']
        product.loc[i,'Sales']=product.loc[i,'Quantity']
        product.loc[i,'Quantity']=product.loc[i,'Discount']
        product.loc[i,'Discount']=product.loc[i,'Profit']
        product.loc[i,'Profit']=product.loc[i,'a']
product=product.drop(labels='a',axis=1)
```

## Rename Column Name For MySQL

```python
colu=[]
for i in product.columns:
    colu.append(i.replace(' ','_'))
```

```python
product.columns=colu
```

## Drop 'Product Name' Column

```python
product=product.drop(labels='Product_Name',axis=1)
```

# Save Fixed Data

```python
product.to_csv('./all_clean_product.csv', index=False, encoding='utf-8')
```
